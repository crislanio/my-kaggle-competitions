## <a name="abstract">Titanic Machine Learning From Disaster</a>

Abstract The sinking of the RMS Titanic is one of the most infamous shipwrecks in history. On April 15, 1912, during her maiden voyage, the Titanic sank after colliding with an iceberg, killing 1502 out of 2224 passengers and crew. This sensational tragedy shocked the international community and led to better safety regulations for ships.
One of the reasons that the shipwreck led to such loss of life was that there were not enough lifeboats for the passengers and crew. Although there was some element of luck involved in surviving the sinking, some groups of people were more likely to survive than others, such as women, children, and the upper-class.
In this challenge, we ask you to complete the analysis of what sorts of people were likely to survive. In particular, we ask you to apply the tools of machine learning to predict which passengers survived the tragedy.

```julia
train = readtable("data/train_v3.tsv")
test = readtable("data/test_v3.tsv")
groundtruth = readtable("data/groundtruth_v3.tsv")
gt_test = join(groundtruth, test[:, [:Name, :PassengerId]], on=:Name);
```

```julia
features = setdiff(names(test), [:Name, :PassengerId])
feature_space = gen_feature_space(features)
label = setdiff(names(train), names(test))[1];
```

## Train Models

### XGBoost

```julia
num_rounds = 10
watchlist  = [(dtrain, "train"), (dval, "eval")]
metrics    = ["error"]
params     = Dict("objective"         => "binary:logistic",
                   "booster"          => "gbtree",
                   "eta"              => .1,
                   "alpha"            => .5,
                   "gamma"            => .0,
                   "max_depth"        =>  5,
                   "colsample_bytree" => .5,
                   "min_child_weight" =>  10,
                   "subsample"        => .5,
                   "seed"             =>  1)

println("Training Base Model...")
tic()
model_xgb = XGBoost.xgboost(dtrain, num_rounds, param=params, metrics=metrics, watchlist=watchlist)
toc()
```

### GLMs

```julia
formulas = gen_formulas(features, feature_space, label, 10)
model_glms = gen_glm(train, formulas, Binomial(), LogitLink());
```

### GLMNets

```julia
model_lassos = train_glmnet(train_x, train_y, Binomial(), 1)
model_glmnets = train_glmnet(train_x, train_y, Binomial(), 0.5)
model_rigdes = train_glmnet(train_x, train_y, Binomial(), 0);
```

### Decision Trees, Random Forests and Adaboost Trees

```julia
model_decision_tree = train_decision_tree(train_x, train_y)
model_random_forest = train_random_forest(train_x, train_y, random_features=500, num_trees=10, portion_samples=.5)
model_adaboost_trees, adaboost_trees_coefs = train_adaptive_boosted_trees(train_x, train_y, num_iteration=10);
```

### Fast Factorization Matrix

```julia
model_ffm = train_fmclassification_sgd(train_x, transform_elements!(copy(train_y), y -> y==0? -1 : 1), rank=2);
```

### Scikit-Models

```julia
models_scikit = scikit_all_classifier_models
train_scikit_models(scikit_all_classifier_models, train_x, train_y);
```

```julia
yhats_train   = predict_all(train_x, train_vw_file)
yhats_val     = predict_all(val_x, val_vw_file)
yhats_gt      = predict_all(gt_x, gt_vw_file)
yhats_gt_test = predict_all(gt_test_x, gt_test_vw_file);
```

### Evaluate Validation Data

```julia
models_name       = collect(keys(yhats_train))
precision_train   = [eval_precision(train_y, predictions) for (key, predictions) in yhats_train]
precision_val     = [eval_precision(val_y, predictions) for (key, predictions) in yhats_val]
precision_gt      = [eval_precision(gt_y, predictions) for (key, predictions) in yhats_gt]
precision_test_gt = [eval_precision(gt_test_y, predictions) for (key, predictions) in yhats_gt_test]
results = vcat(precision_train, precision_val, precision_gt, precision_test_gt)
groups_name = ["Train", "Val", "GT", "GT Test"]
results = gen_dataframe_results(models_name, results, groups_name)
sort!(results, cols=[:Model]);
```

```julia
results_gt_val = results[results[:Phase] .== "GT Test", :]
results_gt_val = results_gt_val[results_gt_val[:Result] .> .75, :]
selected_models = collect(Set(sort(results_gt_val, cols=[:Result], rev=[true])[:Model]))[1:50]
results = results[Vector{Bool}([m in selected_models for m in results[:Model]]), :]
results = results[Vector{Int}(results[:Phase] .== "Val") + Vector{Int}(results[:Phase] .== "GT Test") .> 0, :]
sort!(results, cols=[:Result, :Model], rev=[true, true]);
```
